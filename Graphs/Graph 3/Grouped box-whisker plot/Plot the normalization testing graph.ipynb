```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("1_test.csv")

sns.boxplot(data=data,x="Feature name",y="Feature Value",hue="type")
```

```python
# normalization
# https://www.delftstack.com/zh/howto/python-pandas/pandas-normalize/
def minmax_norm(df):
    return (df - df.min()) / ( df.max() - df.min())
def mean_norm(df_input):
    return df_input.apply(lambda x: (x-x.mean())/ x.std(), axis=0)

def index_extraction(df):
    a = df["Feature name"].value_counts()
    b = a.index
    return b

Features_name = index_extraction(data)

def data_processing(data,Features_name):
    data_new = pd.DataFrame()
    for i in Features_name:
        mean = data.loc[(data["type"]=="PD")&(data["Feature name"]==i)].mean()
        max = data.loc[(data["type"]=="PD")&(data["Feature name"]==i)].max(numeric_only=True)
        min = data.loc[(data["type"]=="PD")&(data["Feature name"]==i)].min(numeric_only=True)

        a = data.loc[(data["type"]=="PD")&(data["Feature name"]==i)].apply(lambda x: (x-min)/(max-min))
        data_new = pd.concat([data_new,a],axis=0)
    for i in Features_name:
        mean = data.loc[(data["type"]=="Intact")&(data["Feature name"]==i)].mean()
        max = data.loc[(data["type"]=="Intact")&(data["Feature name"]==i)].max(numeric_only=True)
        min = data.loc[(data["type"]=="Intact")&(data["Feature name"]==i)].min(numeric_only=True)

        a = data.loc[(data["type"]=="Intact")&(data["Feature name"]==i)].apply(lambda x: (x-min)/(max-min))
        data_new = pd.concat([data_new,a],axis=0)
    return data_new
a = data_processing(data,Features_name=Features_name)
a
data
```
